# Retrieving and Plotting Global Ocean Data

```python
from ddsapi import Client
```

```python
dataset = 'cglorsv7'
request = {
    'product_type': 'monthly',
    'variable': 'sea_water_potential_temperature',
    'depth': [
        0.5057600140571594,
        5.140361309051514,
        11.773679733276367,
        22.75761604309082
    ],
    'time': {
        'year': '2005',
        'month': ['5', '6']
    },
    'area': {
        'north': 48,
        'south': 30,
        'east': 38,
        'west': -10
    },
    'format': 'netcdf'
}

dcube = Client(quiet=True).retrieve(dataset, request)
```

```python
temperature = dcube['sea_water_potential_temperature']
temperature
```

```python
temperature.plot(figsize=(17, 10), robust=True)
```

```python
temperature.plot(features=('coastline', 'borders', 'lakes', 'rivers'))
```
